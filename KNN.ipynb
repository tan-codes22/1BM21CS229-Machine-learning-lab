```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
iris_data = pd.read_csv('Iris.csv')
iris_data.head()
```

```python
label_mapping = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}
iris_data['Species'] = iris_data['Species'].map(label_mapping)
```

```python
plt.scatter(iris_data['SepalLengthCm'], iris_data['SepalWidthCm'], c=iris_data['Species'],  cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Iris Dataset (Sepal Features)')
plt.colorbar(label='Species')
plt.show()
```

```python
iris_data['Species'].unique()
```

```python
iris_data.drop('Id',inplace=True,axis=1)
X = iris_data.drop('Species', axis=1)
y = iris_data['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5) 
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
